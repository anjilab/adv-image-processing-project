```python
import os
import random
import matplotlib.pyplot as plt
import PIL 
```

```python
epsilons = [0.01, 0.03, 0.1, 0.5]
# plot_selected_files =  os.listdir(os.path.join(cfg["data_root"], 'kvasir_polyp/images')) # 1000
bird_image_adv = os.listdir(os.path.join('/home/anjilabudathoki/dip-project/project-2025/src/datasets/CUB_processed_adversarial_jointmdl_attr_save', "0.01")) # 100
random_img_bird_adv = random.sample(bird_image_adv, 5)


fig, axs = plt.subplots(len(random_img_bird_adv), 4  , figsize=(5,5))
for idx, filename in enumerate(random_img_bird_adv):
    img_list = [os.path.join('/home/anjilabudathoki/dip-project/project-2025/src/datasets/CUB_processed_adversarial_jointmdl_attr_save', f"{epsilon}", filename) for epsilon in epsilons]
   
    
    for col, img_path in enumerate(img_list):
    # Load the image
        if os.path.exists(img_path):
            img = PIL.Image.open(img_path)
        else:
            img = None  # Handle missing files (optional)
            print(f"File not found: {img_path}")
            

        # Plot the image
        if img is not None:
            axs[idx, col].imshow(img)
            axs[idx, col].axis('off')  # Hide axes
            if idx == 0:
                axs[idx, col].set_title(f"ε={epsilons[col]}")
        else:
            axs[idx, col].axis('off')  # Keep the layout consistent

# Adjust layout and show the plot
# plt.subplots_adjust(wspace=0.8, hspace=0.1)
plt.tight_layout()
plt.show()
```
